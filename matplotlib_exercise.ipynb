## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist_data = pd.read_csv("data/hist_data.csv")
```

```python
df_hist_data.head()
```

```python
# aデータとbデータを重ねたヒストグラムの作成
plt.hist(df_hist_data["aデータ"], bins=30, alpha=0.5, color= "orange", label = "aデータ")
plt.hist(df_hist_data["bデータ"], bins=30, alpha=0.5, color = "c", label = "bデータ")

# タイトル、x軸、y軸
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布")

# ラベル(凡例)を表示
plt.legend()

# プロット表示
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter_data = pd.read_csv('data/scatter_data.csv')
```

```python
df_scatter_data.head() # 確認
```

```python
class_0 = df_scatter_data[df_scatter_data["class"] == 0]
class_1 = df_scatter_data[df_scatter_data["class"] == 1]
class_2 = df_scatter_data[df_scatter_data["class"] == 2]

# クラスごとに色を分けて散布図を作成
plt.scatter(class_0["length (cm)"],class_0["width (cm)"], color = "r", label = "class0")
plt.scatter(class_1["length (cm)"],class_1["width (cm)"], color = "b", label = "class1")
plt.scatter(class_2["length (cm)"],class_2["width (cm)"], color = "g", label = "class2")

# タイトル、x軸、y軸
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.title("クラス０～２のlengthとwidthに関する散布図")

# ラベルの呼び出し
plt.legend()

# グリッド線
plt.grid(True)

# プロット表示
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line_data = pd.read_csv('data/line_data.csv')
```

```python
# -------折れ線グラフの作成-------
plt.plot(df_line_data["期間"], df_line_data["件数"], marker = "o")

#x軸の目盛りを斜めにする
plt.xticks(rotation=30)

#タイトル、x軸、y軸
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")

#グリッド線を表示
plt.grid(True)

#プロット表示
plt.show()

# -------棒グラフの作成-------
plt.bar(df_line_data["期間"], df_line_data["費用"])

#x軸の目盛りを斜めにする
plt.xticks(rotation=30)

#タイトル、x軸、y軸
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")

#グリッド線を表示
plt.grid(True)

#プロット表示
plt.show()
```
